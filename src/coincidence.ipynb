```python
import ROOT
import histogram_manager as hm

%jsroot on
```

```python
hists = hm.HistogramManager("../output/stage0_output.root")
```

```python
hEn = hists.get("hEn_d1")
hEn_coin0 = hists.get("hEn_coin0_d1")
hEn_coin1 = hists.get("hEn_coin1_d1")

hEn.Rebin(100)
hEn_coin0.Rebin(100)
hEn_coin1.Rebin(100)
```

```python
hEn_coin0_rate = hEn_coin0/hEn
hEn_coin1_rate = hEn_coin1/hEn
```

```python
c = ROOT.TCanvas("","",600,800)
c.Divide(1,2)

ROOT.gStyle.SetOptStat(0)

c.cd(1)

hEn_coin0_rate.Draw()
hEn_coin1_rate.Draw("same")

hEn_coin1_rate.SetLineColor(ROOT.kRed)
hEn_coin0_rate.GetYaxis().SetTitle("rate")
hEn_coin0_rate.SetTitle("^{139}La Coincidence Comparison")
hEn_coin0_rate.GetXaxis().SetRangeUser(0.5,5)
hEn_coin0_rate.GetYaxis().SetRangeUser(0,1)

leg = ROOT.TLegend(.2,.13)
leg.AddEntry(hEn_coin0_rate,"coin0 rate","L")
leg.AddEntry(hEn_coin1_rate,"coin1 rate","L")
leg.Draw()

c.cd(2)

hEn.Draw()
hEn.GetXaxis().SetRangeUser(0.5,5)


c.Draw()
```

```python
hEgam = hists.get("hEgam_d1")
hEgam_coin0 = hists.get("hEgam_coin0_d1")
hEgam_coin1 = hists.get("hEgam_coin1_d1")
```
