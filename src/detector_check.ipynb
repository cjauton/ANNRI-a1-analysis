```python
i = 0

pos_num_glob = []
entries_glob = []
pos_time_sec_glob = []
channel_glob = []

for file_name in root_files:
    root_hm = hm.HistogramManager(file_name)
    hists = root_hm.get('hEgam')
    
    ch = 0
    for key, hist in hists.items():
        entries = hist.GetEntries()
        
        pos_num_glob.append(pos_num[i])
        entries_glob.append(entries)
        pos_time_sec_glob.append(pos_time_sec[i])
        channel_glob.append(ch)
        
        ch+=1

    i+=1
        
```

```python
df = pd.DataFrame({'position':pos_num_glob,'channel':channel_glob,'entries':entries_glob,'time':pos_time_sec_glob})

df['rate'] = df['entries']/df['time']
```
