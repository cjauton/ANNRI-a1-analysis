```python
import numpy as np
import pandas as pd
```

```python
det = np.array([1,2,3,4,5,6,7,18,19,20,25,26,27,28])

ph_511  = np.array([482,152,151,479,499,514,503,497,518,498,482,552,505,503])
ph_1779 = np.array([1682,531,526,1670,1739,1791,1753,1744,1806,1738,1693,1927,1763,1759])
ph_7724 = np.array([7302,2307,2287,7251,7552,7780,7619,7578,7845,7553,7358,8372,7641,7630])


# df = pd.DataFrame({"det":det,"ph_511":ph_511,"ph_1368":ph_1368,"ph_2754":ph_2754})
# df

slope = []
offset=[]
for i in range(len(det)):
    poly = np.polynomial.Polynomial.fit([ph_511[i],ph_1779[i],ph_7724[i]],[511,1779,7724],deg = 1)
    # poly = np.polynomial.Polynomial.fit([ph_511[i],ph_9034[i]],[511,9034],deg = 1)
    # res = np.polynomial.Polynomial.fit(np.array([0,1]),np.array([0,1]),deg = 1)
    slope.append(poly.convert().coef[1])
    offset.append(poly.convert().coef[0])
    

df_results = pd.DataFrame({"det":det,"slope":slope,"offset":offset})
df_results
```
