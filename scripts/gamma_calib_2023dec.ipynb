```python
import numpy as np
import pandas as pd
```

```python
det = np.array([1,2,3,6,7,9,10,11,13,15,17,18,19,20,22,25,26,27,28])

ph_511  = np.array([483,  543,445,  516,  592,559,503,566,355,498,503,525,557,500,337,486,554,347,326])
ph_1779 = np.array([1684,1892,1552,1797, 2056,1949,1754,1971,1235,1736,1753,1835,1941,1741,1176,1693,1931,1107,1142])
ph_7724 = np.array([7315,8215,6743,7802, 8936,8466,7616,8563,5367,7540,7616,7971,8430,7564,5106,7352,8388,4800,4950])
ph_9034 = np.array([8530,9592,7890,9135,10498,9905,8912,10021,6279,8821,8915,9333,9868,8854,6050,8606,9820,5613,5784])

# df = pd.DataFrame({"det":det,"ph_511":ph_511,"ph_1368":ph_1368,"ph_2754":ph_2754})
# df

slope = []
offset=[]
for i in range(len(det)):
    poly = np.polynomial.Polynomial.fit([ph_511[i],ph_1779[i],ph_7724[i],ph_9034[i]],[511,1779,7724,9034],deg = 1)
    poly = np.polynomial.Polynomial.fit([ph_511[i],ph_9034[i]],[511,9034],deg = 1)
    # res = np.polynomial.Polynomial.fit(np.array([0,1]),np.array([0,1]),deg = 1)
    slope.append(poly.convert().coef[1])
    offset.append(poly.convert().coef[0])
    

df_results = pd.DataFrame({"det":det,"slope":slope,"offset":offset})
df_results
```

```python
df_results.to_csv("../calib/calib_2023dec.csv",index=False)
```
