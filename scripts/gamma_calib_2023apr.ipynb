```python
det = np.array([1,2,3,6,7,9,10,11,13,15,17,18,19,20,25,26])
ph_511 = np.array([483,543,445,516,593,559,498,555,354,498,503,526,555,499,487,547])
ph_1368 = np.array([1296,1459,1194,1383,1582,1500,1338,1489,951,1336,1349,1412,1488,1339,1310,1468])
ph_2754 = np.array([2609,2936,2403,2783,3184,3020,2695,2997,1915,2690,2715,2843,2995,2697,2636,2954])

# df = pd.DataFrame({"det":det,"ph_511":ph_511,"ph_1368":ph_1368,"ph_2754":ph_2754})
# df

slope = []
offset=[]
for i in range(len(det)):
    poly = np.polynomial.Polynomial.fit([ph_511[i],ph_1368[i],ph_2754[i]],[511,1368,2754],deg = 1)
    # res = np.polynomial.Polynomial.fit(np.array([0,1]),np.array([0,1]),deg = 1)
    slope.append(poly.convert().coef[1])
    offset.append(poly.convert().coef[0])
    

df_results = pd.DataFrame({"det":det,"slope":slope,"offset":offset})
df_results
```
